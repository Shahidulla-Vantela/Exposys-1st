### **HANDLING OUTLIERS**

```python
q1 = df['Profit'].quantile(0.25)
q3 = df['Profit'].quantile(0.75)
iqr = q3 - q1
print(iqr)
```

```python
ub = q3 + 1.5 * iqr
lb = q1 - 1.5 * iqr
print(ub)
print(lb)
```

```python
upper_array = np.where(df['Profit']>=ub)[0]
lower_array = np.where(df['Profit']<=lb)[0]
```

```python
df.drop(index = upper_array, inplace = True)
df.drop(index = lower_array, inplace = True)
```

```python
y = df['Profit']
X = df.drop(columns = 'Profit', axis =1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(X)
X = pd.DataFrame(X, columns = ['R&D Spend','Administration','Marketing Spend'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
ypred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print(mean_squared_error(y_test, ypred))
print(mean_absolute_error(y_test, ypred))
print('R2 score : ', r2_score(y_test, ypred))
```

```python
lr.coef_
```

```python
pd.DataFrame(lr.coef_,index = X.columns, columns=['Coefficients'])
```

```python
tr = lr.score(X_train, y_train)
te = lr.score(X_test, y_test)
print('Training Accuracy :', tr)
print('Testing Accuracy :', te)
```

```python
sns.distplot(y_test - ypred)
```

```python
plt.scatter(y_test, ypred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfg = RandomForestRegressor(n_estimators = 300, criterion = 'squared_error', random_state = 18)
```

```python
rfg.fit(X_train, y_train)
ypred_rfg = rfg.predict(X_test)
ypred_rfg
```

```python
print('R2 Score of RFG is : ', r2_score(y_test, ypred_rfg))
```

```python
tr2 = rfg.score(X_train, y_train)
te2 = rfg.score(X_test, y_test)
print('Training Accuracy :', tr2)
print('Testing Accuracy :', te2)
```

```python
plt.scatter(y_test, ypred_rfg)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(n_estimators=90, criterion='squared_error',learning_rate=0.01,loss='squared_error')
```

```python
gbr.fit(X_train, y_train)
ypred_gbr = gbr.predict(X_test)
ypred_gbr
```

```python
print('R2 score of GBR is : ', r2_score(y_test, ypred_gbr))
```

```python
tr3 = gbr.score(X_train, y_train)
te3 = gbr.score(X_test, y_test)
print('Training Accuracy :', tr3)
print('Testing Accuracy :', te3)
```

```python
plt.title('Gradient Boosting Model')
plt.scatter(y_test, ypred_gbr)
plt.show()
```
